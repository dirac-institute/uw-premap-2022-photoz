```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ellip_template = np.genfromtxt('data/sed_templates/CWW_KINNEY/CWW_E_ext.sed', unpack = True)
ellip_template
```

```python
plt.plot(ellip_template[0], ellip_template[1])
plt.xlabel("Wavelengths")
plt.ylabel("Flux")
```

```python
irregular = np.genfromtxt('data/sed_templates/CWW_KINNEY/CWW_Im_ext.sed', unpack = True)
spiral = np.genfromtxt('data/sed_templates/CWW_KINNEY/CWW_Sbc_ext.sed', unpack = True)
```

```python
wavelength_emit = ellip_template[0]/10

z = 0.1
plt.plot(wavelength_emit*z+wavelength_emit, ellip_template[1]/np.sum(ellip_template[1]), label = "Elliptical z=0.1")

z = 0.0
plt.plot(wavelength_emit*z+wavelength_emit, ellip_template[1]/np.sum(ellip_template[1]), label = "Elliptical z=0.0")

plt.xlabel("Wavelengths (Nm)")
plt.ylabel("Flux")
plt.xlim(0,3000) #this expands the graph out to see it better
plt.legend()
```

```python
spiral = np.genfromtxt('data/sed_templates/CWW_KINNEY/CWW_Sbc_ext.sed', unpack = True)
```

```python
plt.plot(ellip_template[0]/10, ellip_template[1]/np.sum(ellip_template[1]), label = "Elliptical")
plt.plot(spiral[0]/10, spiral[1]/np.sum(spiral[1]), label = "Spiral")
plt.plot(irregular[0]/10, irregular[1]/np.sum(irregular[1]), label = "Irregular")

plt.xlim(0,4000)
plt.xlabel("Wavelengths (Nm)")
plt.ylabel("Flux")
plt.legend()
```

```python
total_g = np.genfromtxt('data/lsst_baseline_throughputs/total_g.dat', unpack = True)
total_i = np.genfromtxt('data/lsst_baseline_throughputs/total_i.dat', unpack = True)
```
